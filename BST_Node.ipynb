```python
# 이진탐색트리 탐색 (작으면 왼쪽 서브트리, 크면 오른쪽 서브트리)
def search_bst(n,key):
    if n == None:
        return None
    elif key == n.key:
        return n
    elif key < n.key:
        return search_bst(n.left,key)
    else:
        return search_bst(n.right,key)
```

```python
# 최대 최소 노드
def search_max_bst(n):
    while n != None and n.right != None: #최대 탐색이므로 오른쪽으로
        n = n.right
    return n

def search_min_bst(n):
    while n != None and n.left != None: 
        n = n.left
    return n
```

```python
#이진탐색트리 삽입
def insert_bst(r,n): #r = 루트 노드
    if n.key < r.key:
        if r.left is None: # r보다 작고 r 왼쪽이 None 이면 그냥 삽입
            r.left = n
            return True
        else:
            return insert_bst(r.left,n)
    elif n.key > r.key:
        if r.right is None:
            r.right = n
            return True
        else:
            return insert_bst(r.right, n)
    else:
        return False
        
```

```python
#이진탐색트리 삭제연산 case 1(단말노드 삭제), case 2(자식이 하나인 노드), case 3(자식 2개)
def delete_bst_case1(parent,node,root):
    if parent is None:
        root = None
    else:
        if parent.left == node:
            parent.left = None
        else:
            parent.right = None
    
    return root

def delete_bst_case2(parent,node,root): #node = 삭제할 node
    if node.left is not None:
        child = node.left
    else:
        child = node.right
    
    if node == root: #루트 제거
        root = child
    else:
        if node is parent.left: #parent는 node의 부모
            parent.left = child
        else:
            parent.right = child
    
    return root

def delete_bst_case3(parent,node,root): #node : 삭제할 node / succp : 후계자 부모 / succ : 후계자
    succp = node
    succ = node.right
    while (succ.left != None):
        succp = succ
        succ = succ.left #후계자의 오른쪽 자식을 후계자 위치로 바꿈
        
    if (succp.left == succ):
        succp.left = succ.right
    else:
        succp.right = succ.right
    node.key = succ.key
    node.value = succ.value
    node = succ;
    
    return root #실제 출력값은 node가 사라지지 않지만 링크는 바뀜
        
```

```python
def delete_bst(root,key):
    if root == None: return None

    parent = None
    node = root
    while node != None and node.key != key: #삭제할 노드 탐색
        parent= node
        if key < node.key: node = node.left
        else: node = node.right;
    
    if node == None: return None #삭제할 노드 없음
    if node.left == None and node.right == None: #단말노드 삭제
        root = delete_bst_case1(parent,node,root)
    elif node.left == None or node.right == None: #자식 1개
        root = delete_bst_case2(parent,node,root)
    else:
        root = delete_bst_case3(parent,node,root)
    return root
        
        
```

```python
#LL회전
def rotateLL(A):
    B = A.left
    A.left = B.right
    B.right = A
    return B

def rotateRR(A):
    B = A.right
    A.right = B.left
    B.left = A
    return B
```
